```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import plotly.tools as tls
market_datajson = json.load(open(
  'C:/Users/ext.j.g.berenguel/jgonzal1.github.io/data/json/marketstack.json'
))
market_data_df = pd.json_normalize\
  (market_datajson, record_path=["data"])[["date", "close", "name"]]\
  .pivot(index="date", columns="name", values="close").round(3)
market_data_df.index = pd.to_datetime(market_data_df.index).date
# If any of the first row is empty, then drop it
if market_data_df.iloc[0].isnull().any():
  market_data_df = market_data_df.drop(market_data_df.index[0])
display(market_data_df)
fig, ax = plt.subplots()
for col in list(market_data_df.columns):
  market_data_df[f"{col}_pct_change"] =\
    (market_data_df[col] / market_data_df[col].iloc[0] - 1) * 100
  try:
    ax.plot(
      market_data_df.index,
      market_data_df[f"{col}_pct_change"].values,
      label=col
    )
    ax.set_ylabel("Percentage Change (%)")
    # ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))
    # Show day of the week on x axis
    # ax.xaxis.set_major_formatter(plt.FuncFormatter(
    # lambda x, _: pd.to_datetime(x).strftime('%Y-%m-%d')
    # ))
  except Exception as e:
    print(e)
    continue
ax.grid(True)
ax.legend()
plotly_fig = tls.mpl_to_plotly(fig)
plotly_fig.show()
```

```python
const fs = require('fs');

function readJsonFile(filePath) {
  try {
    const data = fs.readFileSync(filePath, 'utf8');
    const jsonData = JSON.parse(data);
    console.log('JSON Data:', jsonData);
    return jsonData;
  } catch (error) {
    console.error('Error reading or parsing the file:', error.message);
  }
}

const filePath = 'C:/Users/ext.j.g.berenguel/jgonzal1.github.io/'+
  'data/json/marketstack.json';
readJsonFile(filePath);
```
